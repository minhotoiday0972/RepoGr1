```python
import pyodbc
class Job:
    def __init__(self,jobName,jobID,salary,yearsOfExperience):
        self.jobName=jobName
        self.jobID=jobID
        self.salary=salary
        self.yearsOfExperience=yearsOfExperience
    def showJobInfo(self):
        print("jobName: ",self.jobName,", jobID: ",self.jobID,", salary: ",self.salary,", yearsOfExperience: ",self.yearsOfExperience)
    def calculateSalary(self,bonus):
        print("sumSalary: ",self.salary+bonus)
class Person:
    def __init__(self,name,age,address):
        self.name=name
        self.age=age
        self.address=address
    def showPersonalInfo(self):
        print("name: ",self.name,", age: ",self.age,", address: ",self.address)

class Doctor(Job,Person):
    def __init__(self,name, age, address,jobName,jobID,salary,yearsOfExperience,specialization):
        Job.__init__(self, jobName, jobID, salary, yearsOfExperience)
        Person.__init__(self, name, age, address)
        self.specialization=specialization
    def showJobInfo(self):
        super().showJobInfo()
        print("specialization: ",self.specialization)
class Engineer(Job,Person):
    def __init__(self,name, age, address,jobName,jobID,salary,yearsOfExperience,field):
        Job.__init__(self, jobName, jobID, salary, yearsOfExperience)
        Person.__init__(self, name, age, address)
        self.field=field
    def showJobInfo(self):
        super().showJobInfo()
        print("field: ",self.field)
class Teacher(Job,Person):
    def __init__(self,name, age, address,jobName,jobID,salary,yearsOfExperience,subject):
        Job.__init__(self, jobName, jobID, salary, yearsOfExperience)
        Person.__init__(self, name, age, address)
        self.subject=subject
    def showJobInfo(self):
        super().showJobInfo()
        print("subject: ",self.subject)

doctor = Doctor("Alice", 35, "123 Main St", "Doctor", "D01", 30000, 10, "Cardiology")
engineer = Engineer("Bob", 28, "456 Elm St", "Engineer", "E01", 25000, 5, "Software")
teacher = Teacher("Charlie", 40, "789 Oak St", "Teacher", "T01", 20000, 15, "Mathematics")

# Hiển thị thông tin và tính lương
#doctor.showPersonalInfo()
#doctor.showJobInfo()
#doctor.calculateSalary(100)

#Yêu cầu 2
def chooseJob(JobL):
    
    LJob_dcgioithieu=[]
    SL=int(input("nhập vào số lượng ứng viên cần giới thiệu:"))
    Doctors=sorted([j for j in JobL if isinstance(j,Doctor)],key=lambda x: (-x.yearsOfExperience))
    print("Cac ung vien duoc chon")
    while SL>0 and Doctors:
        LJob_dcgioithieu.append(Doctors.pop(0))
        SL-=1
    if SL>0:
        Engineers=sorted([j for j in JobL if isinstance(j,Engineer)],key=lambda x: (-x.yearsOfExperience,x.field))
        while SL>0 and Engineers:
            LJob_dcgioithieu.append(Engineers.pop(0))
            SL-=1
    if SL>0:
        Teachers=sorted([j for j in JobL if isinstance(j,Teacher)],key=lambda x: (-x.yearsOfExperience))
        while SL>0 and Teachers:
            LJob_dcgioithieu.append(Teachers.pop(0))
            SL-=1
    for i in LJob_dcgioithieu:
        i.showPersonalInfo()
JobL=[
    Doctor("Alice", 35, "123 Main St","BSTimMach","M01",20000000,3,"Da khoa"),
    Doctor("Bob", 28, "456 Elm St","BSSanKhoa","N01",18000000,2,"Nhi khoa"),
    Engineer("Charlie", 40, "789 Oak St","Data Engineer","D02",25000000,1,"Data"),
    Engineer("Cavid", 30, "101 Pine St","Electrical Engineer","C02",22000000,3,"Electricity"),
    Teacher("Eva", 45, "202 Maple St", "highschool teacher","H01",12000000,5,"Mathemetics"),
    Teacher("Frank", 38, "303 Birch St","Associate Professor","P01",24000000,7,"English Speaking")
]
chooseJob(JobL)
#Yêu cầu 3
def save_JobL_data_to_file():
    try:
        with open("job_list.csv", "w", encoding="utf-8") as file:
            file.write("Danh sách nghề nghiệp:\n")
            for job in JobL:
                # Ghi thông tin  vào file
                file.write(f"jobType: {type(job).__name__}, ")
                file.write(f"name : {job.name}, ")
                file.write(f" age: {job.age}, ")
                file.write(f"address: {job.address}, ")
                file.write(f"Tên job: {job.jobName}, ")
                file.write(f"jobID: {job.jobID}, ")
                file.write(f"salary: {job.salary}, ")
                file.write(f"experience: {job.yearsOfExperience}, ")
            
                if isinstance(job, Doctor):
                    file.write(f"specialization: {job.specialization}\n")
                elif isinstance(job, Engineer):
                    file.write(f" field: {job.field}\n")
                elif isinstance(job, Teacher):
                    file.write(f"subject: {job.subject} \n")
                else:# ket thuc for , het lenh nen with dong file tu dong
                    file.write("\n")#chỉ ghi một dòng trống.
        print("Đã lưu thông tin phương tiện vào file 'job_list.csv'.")
    except Exception as e:#(ví dụ: quyền truy cập file),
        print(f"Đã xảy ra lỗi khi lưu file: {e}")
save_JobL_data_to_file()


def load_JobL__data_from_file():
    JobLL=[]
    try:
        with open("job_list.csv", "r", encoding="utf-8") as file:
            lines = file.readlines()
        for line in lines[1:]:  
            job_info = line.strip().split(", ")

            job_type = job_info[0].split(": ")[1]  
            name = job_info[1].split(": ")[1]
            age = int(job_info[2].split(": ")[1])
            address = job_info[3].split(": ")[1]
            jobName = job_info[4].split(": ")[1]          
            jobID = job_info[5].split(": ")[1]    
            salary= int(job_info[6].split(": ")[1])         
            yearsOfExperience = int(job_info[7].split(": ")[1])     # Năm sản xuất
            #do doc la string
    
    
            if job_type == "Doctor":
                specialization = job_info[8].split(": ")[1] 
                new_job = Doctor(name, age, address,jobName, jobID, salary, yearsOfExperience, specialization)
            elif job_type == "Engineer":
                field = job_info[8].split(": ")[1] 
                new_job = Engineer(name, age, address,jobName, jobID, salary, yearsOfExperience, field)
            elif job_type == "Teacher":
                subject = job_info[8].split(": ")[1]
                new_job = Teacher(name, age, address,jobName, jobID, salary, yearsOfExperience,subject)
            else:
                print(f"Loại nghề nghiệp không xác định: {job_type}")
                continue  
            JobLL.append(new_job)

        print("Đã tải thông tin phương tiện từ file 'job_list.csv'.")
        return JobLL
    except Exception as e:
        print(f"Đã xảy ra lỗi khi đọc file: {e}")
load_JobL__data_from_file()

def show_JobL_data_from_file(JobLL):
    JobLL=sorted([j for j in JobL if isinstance(j,Job)],key=lambda x: (-ord(x.name[0]), x.salary))
    for i in JobLL:
        i.showPersonalInfo()
        i.showJobInfo()
#show_JobL_data_from_file(JobLL)

#Yêu cầu 4

def connect_to_db():
    try:
        connection = pyodbc.connect('DRIVER={ODBC Driver 17 for SQL Server};'
                                    'Server=LAB3-MAY19\SQLEXPRESS;'
                                    'DATABASE=Jobs;'
                                    'Trusted_Connection=yes')
                                    
        return connection
    except Exception as e:
        print(f"Lỗi kết nối cơ sở dữ liệu: {e}")
        return None

    
    
    
def add_job_to_db(connection, job):
    try:
        cursor = connection.cursor()
        query = """
INSERT INTO Jobs (JobID, JobName, Salary, Experience, jobType, Specialization, field, Subject)
VALUES (?, ?, ?, ?, ?, ?, ?, ?)
"""
        specialization = (
            job.specialization if isinstance(job, Doctor)
            else job.field if isinstance(job, Engineer)
            else job.subject
        )
        job_type = type(job).__name__
        cursor.execute(query, job.jobID, job.jobName, job.salary, job.yearsOfExperience,  job_type,specialization, None, None)
        connection.commit()
        print("Thêm nghề nghiệp thành công!")
    except Exception as e:
        print(f"Lỗi khi thêm nghề nghiệp: {e}")

def update_job_in_db(connection, job_id, new_salary):
    try:
        cursor = connection.cursor()
        query = "UPDATE Jobs SET Salary = ? WHERE JobID =?"
        cursor.execute(query, new_salary, job_id)
        connection.commit()
        print("Cập nhật lương thành công!")
    except Exception as e:
        print(f"Lỗi khi cập nhật nghề nghiệp: {e}")
def delete_job_from_db(connection, job_id):
    try:
        cursor = connection.cursor()
        query = "DELETE FROM Jobs WHERE JobID = ?"
        cursor.execute(query, job_id)
        connection.commit()
        print("Xóa nghề nghiệp thành công!")
    except Exception as e:
        print(f"Lỗi khi xóa nghề nghiệp: {e}")
def get_all_jobs_from_db(connection):
    try:
        cursor = connection.cursor()
        query = "SELECT * FROM Jobs"
        cursor.execute(query)
        jobs = cursor.fetchall()
        for job in jobs:
            print(job)
    except Exception as e:
        print(f"Lỗi khi truy vấn dữ liệu: {e}")
def filter_jobs_by_criteria(connection, min_salary=0, min_experience=0):
    try:
        cursor = connection.cursor()
        query = """
        SELECT * FROM Jobs
        WHERE Salary >= ? AND Experience >= ?
        """
        cursor.execute(query, min_salary, min_experience)
        jobs = cursor.fetchall()
        for job in jobs:
            print(job)
    except Exception as e:
        print(f"Lỗi khi lọc dữ liệu: {e}")
def menu():
    connection = connect_to_db()
    if not connection:
        return
    
    while True:
        print("\n=== MENU ===")
        print("1. Thêm nghề nghiệp")
        print("2. Cập nhật lương")
        print("3. Xóa nghề nghiệp")
        print("4. Xem danh sách nghề nghiệp")
        print("5. Tìm kiếm nghề nghiệp")
        print("6. Thoát")
        choice = input("Chọn chức năng: ")

        if choice == "1":
            # Thêm nghề nghiệp
            job = Doctor("Alice", 35, "123 Main St", "Cardiologist", "D01", 30000, 10, "Cardiology")
            add_job_to_db(connection, job)

        elif choice == "2":
            # Cập nhật lương
            job_id = input("Nhập JobID: ")
            new_salary = int(input("Nhập lương mới: "))
            update_job_in_db(connection, job_id, new_salary)

        elif choice == "3":
            # Xóa nghề nghiệp
            job_id = input("Nhập JobID để xóa: ")
            delete_job_from_db(connection, job_id)

        elif choice == "4":
            # Xem danh sách
            get_all_jobs_from_db(connection)

        elif choice == "5":
            # Tìm kiếm
            min_salary = int(input("Nhập mức lương tối thiểu: "))
            min_experience = int(input("Nhập số năm kinh nghiệm tối thiểu: "))
            filter_jobs_by_criteria(connection, min_salary, min_experience)

        elif choice == "6":
            print("Thoát chương trình.")
            connection.close()
            break

        else:
            print("Lựa chọn không hợp lệ.")
menu()
```
