```python
from datetime import datetime
from datetime import date
import pandas as pd
from pandas import DataFrame

ngayHienTai = datetime.strptime('07/08/2023', "%d/%m/%Y").date()

class DuAn:
    maDuAn = ''
    tenDuAn = ''
    quanLy = ''
    ngayBatDau = ''
    ngayKetThuc = ''
    ngayHoanThanh = ''
    
    def __init__(self, ma, ten, ql, bd = None, kt= None, ht = None) :
        self.maDuAn = ma
        self.tenDuAn = ten
        self.quanLy = QuanLy(ql.maThanhVien, ql.tenThanhVien)
        self.ngayBatDau = bd
        self.ngayKetThuc = kt
        self.ngayHoanThanh = ht
        self.thongtinduan()
        
    def setMaDuAn(self, ma):
        self.maDuAn = ma
    def setTenDuAn(self, ten):
        self.tenDuAn = ten
    def setMaQuanLy(self, ma):
        self.quanLy  = ma
    def setNgayBatDau (self, bd):
        self.ngayBatDau = bd
    def setNgayKetThuc(self, kt):
        self.ngayKetThuc = kt
    def setNgayHoanThanh(self, ht):
        self.ngayHoanThanh = ht
        
    def getMaDuAn(self):
        self.maDuAn
    def getTenDuAn(self):
        return self.tenDuAn
    def getquanLy (self):
        return self.quanLy
    def getNgayBatDau(self):
        return self.ngayBatDau
    def getNgayKetThuc(self):
        return self.ngayKetThuc
    def getNgayHoanThanh(self):
        return self.ngayHoanThanh
    
    def thongtinduan(self):
        self.quanLy.thongTinDuAn.extend([str(self.maDuAn), str(self.tenDuAn), str(self.ngayBatDau),
                                    str(self.ngayKetThuc)])
        
    
class ThanhVien:
    maThanhVien = ''
    tenThanhVien = ''
    congViec = []
    
    def __init__(self, ma, ten):
        self.maThanhVien = ma
        self.tenThanhVien = ten
        self.congViec = [] # CongViec
        
    def setMaThanhVien(self, ma):
        self.maThanhVien = ma
    def setTenThanhVien (self, ten):
        self.tenThanhVien = ten
    def setCongViec(self, cv):
        self.congViec.append(cv)
        
    def getMaThanhVien(self):
        return self.maThanhVien
    def getTenThanhVien(self):
        return self.tenThanhVien
    def getCongViec(self):
        return self.congViec
    
    def xemDanhSachCongViec(self):
        print('Danh Sach CongViec: ')
        for congViec in self.congViec:
            congViec.getThongTinCongViec()
        print()
    
    def capNhatTrangThaiCongViecDangLam(self, cv, ngayHomNay):
        ngayHomNay = datetime.strptime(ngayHomNay, "%d/%m/%Y").date()
        ngayGiao = datetime.strptime(cv.ngayGiao, "%d/%m/%Y").date() 
        
        if ngayHomNay >= ngayGiao:
            for congviec in self.congViec:
                if congviec.maCongViec == cv.maCongViec:
                    congviec.dangLam = True
                    return
        else: 
            return 'Chua den luc lam'
        return 'Khong tim thay cong viec'
         
        
        
    def capNhatTrangThaiCongViecHoanThanh(self,  cv, ngayhoanThanh):
        for congViec in self.congViec:
            if congViec.maCongViec == cv.maCongViec:
                congViec.hoanThanh = True
                congViec.dangLam = False
                congViec.ngayHoanThanh = ngayhoanThanh
                break
        return 'Khong tim thay cong viec'
                
class QuanLy(ThanhVien):
    maDuAnQuanLy = ''
    danhSachThanhVien = []
    danhSachCongViec = []
    
    def __init__(self, maThanhVien, tenThanhVien):
        super().__init__(maThanhVien, tenThanhVien)
        self.congViec = []
        self.danhSachThanhVien = []
        self.danhSachCongViec = []
        self.thongTinDuAn = []
        
    def getMaDuaAnQuanLy(self):
        return self.maDuAnquanLy
    def getDanhSachCongViec(self):
        return self.danhSachCongViec
    def getDanhSachThanhVien(self):
        return self.danhSachThanhVien
    
    def themCongViec(self, congViec):
        self.danhSachCongViec.append(congViec)
    
    def themThanhVien(self, thanhVien):
        self.danhSachThanhVien.append(thanhVien)
    
    def giaoViec(self, thanhVien, congViec):
        congViec.setThanhVienCongViec(thanhVien)
        thanhVien.setCongViec(congViec)
        
    def canhBaoDenHan(self):
        pass
    
    def congViecChuaHoanThanh(self, ngayHomNay ):
        for cv in self.danhSachCongViec:
            hanChot = datetime.strptime(cv.hanChot, "%d/%m/%Y").date()
            ngayGiao = datetime.strptime(cv.ngayGiao, "%d/%m/%Y").date()
            if cv.dangLam == True:
                if hanChot < ngayHomNay :
                    cv.hoanThanh = False
    
    def xuatBaoCaoCongViecChuaHoanThanh(self):
        baoCao = BaoCao()
        for cv in self.danhSachCongViec:
            if cv.hoanThanh == False :
                baoCao.danhSachMa.append(cv.maCongViec)
                baoCao.danhSachTen.append(cv.tenCongViec)
                baoCao.danhSachNgayGiao.append(cv.ngayGiao)
                baoCao.danhSachHanChot.append(cv.hanChot)
                baoCao.danhSachThanhVienCV.append(cv.thanhVienCongViec.tenThanhVien)
        gop = zip(baoCao.danhSachMa, baoCao.danhSachTen, baoCao.danhSachNgayGiao, baoCao.danhSachHanChot, 
                  baoCao.danhSachThanhVienCV)
        df = DataFrame(gop, columns = ['MaCongViec', 'tenCongViec', 'NgayGiao', 'HanChot', 'ThanhVien'])
        return  df
    
    def xuatBaoCaoCongViecDaHoanThanh(self):
        df = self.xuatBaoCaoTatCaCongViec()
        df1 = df.loc[df['TrangThai'] == 'Hoan Thanh']
        return df1
    
    def xuatBaoCaoCongViecDangLam(self):
        df = self.xuatBaoCaoTatCaCongViec()
        df1 = df.loc[df['TrangThai'] == 'Dang Lam']
        return df1
              
    def xuatBaoCaoTatCaCongViec(self):
        baoCao = BaoCao()
        for cv in self.danhSachCongViec:
            baoCao.danhSachMa.append(cv.maCongViec)
            baoCao.danhSachTen.append(cv.tenCongViec)
            baoCao.danhSachNgayGiao.append(cv.ngayGiao)
            baoCao.danhSachHanChot.append(cv.hanChot)
            baoCao.danhSachThanhVienCV.append(cv.thanhVienCongViec.tenThanhVien)
            if cv.hoanThanh == True and cv.dangLam == False:
                baoCao.trangThaiCongViec.append('Hoan Thanh')
                baoCao.ngayHoanThanh.append(cv.ngayHoanThanh)
            if cv.hoanThanh == False:
                baoCao.trangThaiCongViec.append('chua Hoan Thanh')
                baoCao.ngayHoanThanh.append(cv.ngayHoanThanh)
            if cv.dangLam == True:
                baoCao.trangThaiCongViec.append('Dang Lam')
                baoCao.ngayHoanThanh.append('')
            
        gop = zip(baoCao.danhSachMa, baoCao.danhSachTen, baoCao.danhSachNgayGiao, baoCao.danhSachHanChot, 
                  baoCao.danhSachThanhVienCV, baoCao.trangThaiCongViec, baoCao.ngayHoanThanh)
        df = DataFrame(gop, columns = ['MaCongViec', 'tenCongViec', 'NgayGiao', 'HanChot', 'ThanhVien', 'TrangThai', 
                                       'Ngay HoanThanh'])
        name = 'BaoCaoCongViec' + str(ngayHienTai) + '.csv'
        df.to_csv('BaoCao.csv', index = False)
        return  df
                
                
class BaoCao:
    danhSachMa = []
    danhSachTen = []
    danhSachNgayGiao = []
    danhSachHanChot = []
    danhSachThanhVienCV = []
    trangThaiCongViec = []
    ngayHoanThanh = []
    
        
class CongViec:
    maCongViec = ''
    tenCongViec = ''
    #maDuAnCongViec = ''
    noiDungCongViec = ''
    thanhVienCongViec = ''
    ngayGiao = ''
    hanChot = ''
    ngayHoanThanh = 'Qua Han'
    dangLam = False 
    hoanThanh = False
    
    
    def __init__(self, ma, ten, ngayGiao, hanChot, dangLam = False, hoanThanh = False, thanhVien = None):
        self.maCongViec = ma
        self.tenCongViec = ten
        self.ngayGiao = ngayGiao
        self.hanChot = hanChot
        self.ngayHoanThanh = 'Qua Han'
        self.thanhVienCongViec = thanhVien
        self.noiDungCongViec = ''
        self.dangLam = dangLam
        self.hoanThanh = hoanThanh
        
    def setMaCongViec(self, ma):
        self.maCongViec = ma
    def setTenCongViec(self, ten):
        self.tenCongViec = ten
    def setngayGiao (self, ng):
        self.ngayGiao = ng
    def setHanChot(self, hc):
        self.hanChot = hc
    def setNgayHoanThanh(self, ngayHoanThanh):
        self.ngayHoanThanh = ngayHoanThanh
    def setThanhVienCongViec(self, ma):
        self.thanhVienCongViec = ma
    def setNoiDungCongViec(self, nd):
        self.noiDungCongViec = nd
    
    def getMaCongViec(self):
        return self.maCongViec
    def getTenCongViec(self):
        return self.tenCongViec
    def getNgayGiao(self):
        return self.ngayGiao
    def getDangLam (self):
        return self.dangLam
    def getHoanThanh(self):
        return self.hoanThanh
    
    def getThongTinCongViec(self):
        thongTin = [str(self.maCongViec) ,  str(self.tenCongViec), str(self.ngayGiao),
                         str(self.hanChot),  str(self.noiDungCongViec)]
        print(thongTin)
   

        
#Tao ThanhVien
tv1 = ThanhVien('01', 'Nguyen Van Duc')
tv2 = ThanhVien('02', 'Le Cao')
tv3 = ThanhVien('03', 'Huynh Tho')

#Tao du an
XayNha = DuAn(99, 'Xay Nha', tv1, '13/12/2022', '13/08/2023')
ql = XayNha.quanLy
tv1 = ql
#Tao CongViec
cv1 = CongViec('01', 'Thiet ke', '13/12/2022', '13/02/2023')
cv2 = CongViec('02', 'Thi Cong', '13/02/2023', '13/07/2023')
cv3 = CongViec('03', 'Noi That', '13/07/2023', '13/08/2023')
cv4 = CongViec('04', 'Thiet ke He thong Dien', '13/12/2022', '7/01/2023')
cv5 = CongViec('05', 'Thiet ke San Vuon', '13/12/2022','13/01/2023')
cv6 = CongViec('06', 'Tim Nha cung cap Nguyen vat lieu', '13/12/2022', '20/12/2022')
cv7 = CongViec('07', 'Chon Noi that', '13/02/2023', '13/04/2023')
cv8 = CongViec('08', 'Chon San Nha', '13/05/2023', '30/05/2023')
cv9 = CongViec('09', 'Hoach toan', '13/12/2022', '13/02/2023')
cv10 = CongViec('10', 'Tim hieu Yeu Cau Khach hang', '13/12/2022', '13/01/2023')
cv11 = CongViec('11', 'Ke Toan', '13/12/2022', '13/08/2023')
cv12 = CongViec('12', 'Xay dung Mo hinh Ngoi nha', '01/12/2022', '13/02/2023')
cv13 = CongViec('13', 'Quan ly Cong nhan', '13/02/2023', '13/08/2023')
cv14 = CongViec('14', 'Ban giao', '07/08/2023', '13/08/2023')
cv15 = CongViec('15', 'To chuc Tiec', '10/08/2023', '13/08/2023')

#Them ThanhVien
ql.themThanhVien(tv1)
ql.themThanhVien(tv2)
ql.themThanhVien(tv3)

#Them CongViec
ql.themCongViec(cv1)
ql.themCongViec(cv2)
ql.themCongViec(cv3)
ql.themCongViec(cv4)
ql.themCongViec(cv5)
ql.themCongViec(cv6)
ql.themCongViec(cv7)
ql.themCongViec(cv8)
ql.themCongViec(cv9)
ql.themCongViec(cv10)
ql.themCongViec(cv11)
ql.themCongViec(cv12)
ql.themCongViec(cv13)
ql.themCongViec(cv14)
ql.themCongViec(cv15)

#Giao viec
ql.giaoViec(tv1, cv1)
ql.giaoViec(tv1, cv4)
ql.giaoViec(tv1, cv5)
ql.giaoViec(tv1, cv8)
ql.giaoViec(tv1, cv15)

ql.giaoViec(tv2, cv2)
ql.giaoViec(tv2, cv6)
ql.giaoViec(tv2, cv7)
ql.giaoViec(tv2, cv9)
ql.giaoViec(tv2, cv13)

ql.giaoViec(tv3, cv3)
ql.giaoViec(tv3, cv10)
ql.giaoViec(tv3, cv11)
ql.giaoViec(tv3, cv12)
ql.giaoViec(tv3, cv14)

#Xem danh sach CongViec cua ban than
tv1.xemDanhSachCongViec()

print(ngayHienTai)
print(type(ngayHienTai))


#Cap nhat Trang thai CongViec
#CongViec dangthuchien
tv3.capNhatTrangThaiCongViecDangLam(cv14, '07/08/2023')
tv1.capNhatTrangThaiCongViecDangLam(cv15, '10/08/2023') ########
tv3.capNhatTrangThaiCongViecDangLam(cv3, '13/07/2023')
tv2.capNhatTrangThaiCongViecDangLam(cv2, '13/02/2023')
tv3.capNhatTrangThaiCongViecDangLam(cv11, '13/12/2022')
tv3.capNhatTrangThaiCongViecDangLam(cv10, '13/12/2022')
tv2.capNhatTrangThaiCongViecDangLam(cv6, '13/12/2022')
tv3.capNhatTrangThaiCongViecDangLam(cv12, '01/12/2022')
tv2.capNhatTrangThaiCongViecDangLam(cv9, '13/12/2022')
tv1.capNhatTrangThaiCongViecDangLam(cv1, '13/12/2022')
tv2.capNhatTrangThaiCongViecDangLam(cv7, '13/02/2023')
tv1.capNhatTrangThaiCongViecDangLam(cv8, '13/05/2023')
tv2.capNhatTrangThaiCongViecDangLam(cv13, '13/02/2023')
tv1.capNhatTrangThaiCongViecDangLam(cv5, '13/12/2022')
tv1.capNhatTrangThaiCongViecDangLam(cv4, '13/12/2022')

#CongViec hoan thanh
tv3.capNhatTrangThaiCongViecHoanThanh(cv10, '13/01/2023')
tv2.capNhatTrangThaiCongViecHoanThanh(cv6, '18/12/2022')
tv3.capNhatTrangThaiCongViecHoanThanh(cv12, '10/12/2022')
tv2.capNhatTrangThaiCongViecHoanThanh(cv9, '13/02/2023')
tv1.capNhatTrangThaiCongViecHoanThanh(cv1, '13/02/2023')
#CongViec chuaHoanThanh
#(cv7)
#(cv8)
#(cv13)
#(cv5)
#(cv4)
#ql.congViecChuaHoanThanh(ngayHienTai)



#ql.congViecChuaHoanThanh(ngayHienTai)
#ql.xuatBaoCaoCongViecChuaHoanThanh()

ql.xuatBaoCaoTatCaCongViec()

#ql.xuatBaoCaoCongViecDaHoanThanh()

#ql.xuatBaoCaoCongViecDangLam()


```
